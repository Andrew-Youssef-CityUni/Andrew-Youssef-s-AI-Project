# Lab 06: Infamous XOR (again!)

XOR corresponds to a boolean logical gate with the following logic:

You have two inputs, X_0 and X_1, and an output y.
X_0 and X_1 can take the values 0 or 1.

y will be equal to 1 if X_0 and X_1 are different.
If they have the same value, y will output 0.

This simple problem can't be solved with a single layer perceptron, as a linear decision boundary is not expressive enough.
Historically, this shortcoming of the perceptron was taken as an argument long time ago to dismiss the model of Artificial Neural Network, and lead to the first AI winter.

Actually, we know today that this problem can be easily solved with a Multi-layer perceptron.

In this Lab, we will first generate data, then normalize it, and write a very simple multi-layer perceptron to solve this problem.



## 1 - Data generation

I provide the data generation code.
You have to write a class which ill normalize the data by mean and variance, so that training data has mean 0 and variance 1.

Some details:
- the normalization coefficients should be calculated on the training data and validation data alone.
- once calculated, they can be applied to training, validation and testing data.

This is because, testing data should be as much isolated as possible. You can't assume that you know anything about the testing data while training. In particular, you can't use the testing data to normalize your data.


```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs( n_samples=1000, n_features=2, centers=((-2, -2), (-2, 2),(2, 2),(2, -2)), cluster_std = 1)

y[y==2] = 0
y[y==3] = 1

plt.scatter( X[:,0], X[:,1], c=y)
```

```python
import numpy as np

ratio_valid = 0.2
ratio_test = 0.2
X_valid, X_test, X_train = np.split(X, [int(len(X)*ratio_valid), int(len(X)*(ratio_valid+ ratio_test))])
y_valid, y_test, y_train = np.split(y, [int(len(X)*ratio_valid), int(len(X)*(ratio_valid+ ratio_test))])
    

class MeanVarNormalize():
    
    def __init__(self):
        
        self.mean = 0
        self.std = 0
        
    def fit(self, X):
        
        self.mean = np.mean(X, axis = 0)
        self.std = np.std(X, axis = 0)
        
        print('Mean: ', self.mean, ' - Std :', self.std)
        
    def transform(self, X):
        
        X_normalized = (X - self.mean) / self.std
        
        return X_normalized
    
mv_normalization = MeanVarNormalize()

# We use just the data available for training.
# Alternatively, we could use X_train and X_valid together, but not X_test
mv_normalization.fit( X_train )
X_train_n = mv_normalization.transform(X_train)
X_valid_n = mv_normalization.transform(X_valid)
X_test_n = mv_normalization.transform(X_test)

# We can check whether our normalization works correctly
print( np.mean( X_test_n, axis = 0) )
print( np.var( X_test_n, axis = 0))

# That's close enough, and it doesn't need to be perfect.
plt.scatter( X_test_n[:,0], X_test_n[:,1], c=y_test)
```

# 2 - Multi-layer perceptron

In order to solve XOR, we will build a very simple MLP:
- 2 inputs (X_1 and X_2)
- 2 hidden units 
- 1 output (sigmoid)

You have to write a class for this MLP with:
- the forward pass
- the backward pass
- the update of the weights using SGD
- a stopping criterion

In addition, you can write a subclass to implement a Random Search:
- the forward pass
- a random update of the weight
- keep the weights that lead to the lowest error
- stop after a number of epochs



```python
import random

class MLP():
    
    def __init__(self):
        
        # first hidden unit takes input from input neuron 0 and 1
        self.w_h_0_0, self.w_h_1_0, self.b_h_0 = (random.uniform(-1, 1) for i in range(3) )
        
        # second hidden unit takes input from input neuron 0 and 1
        self.w_h_0_1, self.w_h_1_1, self.b_h_1 = (random.uniform(-1, 1) for i in range(3) )
        
        # output neuron takes input from both hidden neurons
        self.w_o_0, self.w_o_1, self.b_o = (random.uniform(-1, 1) for i in range(3) )
        
    def forward(self, X):
        
        activation_hidden_0 = self.sigmoid(X[:, 0]*self.w_h_0_0 + X[:, 1]*self.w_h_1_0 + self.b_h_0 )
        activation_hidden_1 = self.sigmoid(X[:, 0]*self.w_h_0_1 + X[:, 1]*self.w_h_1_1 + self.b_h_1 )
        
        output = self.sigmoid(activation_hidden_0*self.w_o_0 + activation_hidden_1*self.w_o_1 + self.b_o )
        
        return output, activation_hidden_0, activation_hidden_1
    
    def evaluate(self, X, y):
        
        output, _, _ = self.forward(X)
        pred = 1.0*(output>0.5)      
        accuracy = np.sum(1.0*(pred == y))/len(y)
        return accuracy
    
    def loss(self, y_pred, y):
        return np.mean(-y * np.log(y_pred)- (1.0-y) * np.log(1.0 - y_pred))
    
        
    @staticmethod
    def sigmoid(x):
        return 1/(1+np.exp(-x))
    
mlp = MLP()
print(mlp.evaluate(X_test_n, y_test))
```

```python
class MLP_GD(MLP):
    
    def __init__(self, lr = 0.001):
        
        # We call init of the inherited class
        super().__init__()
        
        self.lr = lr
        self.validation_loss_history = []
        self.training_loss_history = []
        
    def fit(self, X_train, y_train, X_valid, y_valid, epochs):
        
        epoch = 0
        continue_training = True 
        
        while continue_training:
            # One parameter update
            loss_tr = self.update(X_train, y_train)
            self.training_loss_history.append(loss_tr)
        
            # valid loss
            outputs, activ_h_0, activ_h_1 = self.forward(X_valid)
            loss_v = self.loss(outputs, y_valid)
            self.validation_loss_history.append(loss_v)
        
            
            epoch += 1
            
            if epoch > epochs:
                continue_training = False
                
        return 'Done'
            
        
    def update(self, X_train, y_train):
        outputs, activ_h_0, activ_h_1 = self.forward(X_train)
        
        # Output layer
        sigma_z = outputs - y_train
        
        # gradients for output layer
        d_w_o_0 = sigma_z * activ_h_0
        d_w_o_1 = sigma_z * activ_h_1
        d_b_o = sigma_z
                
        # Hidden Layer
        sigma_h_0 = sigma_z * self.w_o_0
        sigma_h_1 = sigma_z * self.w_o_1
        
        # gradients for hidden layer
        d_w_h_0_0 = activ_h_0*(1-activ_h_0)*sigma_h_0 * X_train[:, 0]
        d_w_h_1_0 = activ_h_0*(1-activ_h_0)*sigma_h_0 * X_train[:, 1]
        d_w_h_0_1 = activ_h_1*(1-activ_h_1)*sigma_h_1 * X_train[:, 0]
        d_w_h_1_1 = activ_h_1*(1-activ_h_1)*sigma_h_1 * X_train[:, 1]
        d_b_h_0 = activ_h_0*(1-activ_h_0)*sigma_h_0
        d_b_h_1 = activ_h_1*(1-activ_h_1)*sigma_h_1
            
        # Update
        self.w_h_0_0 -= self.lr*np.sum(d_w_h_0_0)
        self.w_h_1_0 -= self.lr*np.sum(d_w_h_1_0)
        self.b_h_0 -= self.lr*np.sum(d_b_h_0)
        
        self.w_h_0_1 -= self.lr*np.sum(d_w_h_0_1)
        self.w_h_1_1 -= self.lr*np.sum(d_w_h_1_1)
        self.b_h_1 -= self.lr*np.sum(d_b_h_1)
        
        self.w_o_0 -= self.lr*np.sum(d_w_o_0)
        self.w_o_1 -= self.lr*np.sum(d_w_o_1)
        self.b_o -= self.lr*np.sum(d_b_o)
        
        outputs, activ_h_0, activ_h_1 = self.forward(X_train)
        loss = self.loss(outputs, y_train)
        
        return loss
                
        
```

```python
mlp_gd = MLP_GD(lr = 0.001)

mlp_gd.fit(X_train_n, y_train, X_valid_n, y_valid, 10000)

print(mlp_gd.evaluate(X_test_n, y_test))
plt.plot(mlp_gd.training_loss_history, 'r-')
plt.plot(mlp_gd.validation_loss_history, 'b-')

plt.show()
```

It works!

Now let's have fun with a Bruteforce Evolutionary approach.

We just need to change the update rule!


```python
class MLP_Evo(MLP_GD):
    
    def __init__(self):
        super().__init__()
            
    def update(self, X_train, y_train):
        
        outputs, _, _ = self.forward(X_train)
        prev_loss = self.loss(outputs, y_train)
        
        # We copy the parameters in memory:
        mem = self.w_h_0_0, self.w_h_1_0, self.b_h_0, self.w_h_0_1, self.w_h_1_1, self.b_h_1, self.w_o_0, self.w_o_1, self.b_o
        
        # We sample new random parameters:
        
        self.w_h_0_0, self.w_h_1_0, self.b_h_0 = (random.uniform(-1, 1) for i in range(3) )
        self.w_h_0_1, self.w_h_1_1, self.b_h_1 = (random.uniform(-1, 1) for i in range(3) )
        self.w_o_0, self.w_o_1, self.b_o = (random.uniform(-1, 1) for i in range(3) )
        
        # If they work better, we keep them:
        outputs, _, _ = self.forward(X_train)
        new_loss = self.loss(outputs, y_train)
        
        if prev_loss < new_loss:
            self.w_h_0_0, self.w_h_1_0, self.b_h_0, self.w_h_0_1, self.w_h_1_1, self.b_h_1, self.w_o_0, self.w_o_1, self.b_o = mem
            return prev_loss
        else:
            return new_loss
        
```

```python
mlp_evo = MLP_Evo()

mlp_evo.fit(X_train_n, y_train, X_valid_n, y_valid, 50000)

print(mlp_evo.evaluate(X_test_n, y_test))
plt.plot(mlp_evo.training_loss_history, 'r-')
plt.plot(mlp_evo.validation_loss_history, 'b-')
```
