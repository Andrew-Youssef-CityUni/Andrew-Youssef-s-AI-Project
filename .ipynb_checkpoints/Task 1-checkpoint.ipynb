Dijkstra's shortest path algorythm is very similar to the above solution which I did before reading about Dijkistra's algorythm. The idea is to the shortest delay in the two adjacent cells while progressing forwards. I have added an array which prints at the end to show the cells and the time which the algorythm chose. 

```python
def dijkstra(game, previous, cell, endNode):
    startN= cell
    endN = endNode
    #print ('startN ', startN.x, startN.y)
    #print('endNode', endNode.x, endNode.y)
    a = startN.x
    b = startN.y
    r,c = game.shape
    
    #check above
    aboveN = Coordinates(a-1, b)
    #print('aboveN ', aboveN.x, aboveN.y)
    #check below
    belowN = Coordinates(a+1,b)
    #print('belowN ', belowN.x, belowN.y)
    #check front
    frontN = Coordinates(a, b+1)
    #print('frontN ', frontN.x, frontN.y)
    
   
    
    #variables to keep track of best cell with the least time
    timeInCell= 10
    bestMove = startN
    
    #logic for the cells adjacent to the last cell
    
    if belowN.x == endN.x and belowN.y == endN.y:
        #timeInCell= game[startN.x, startN.y]
        bestMove=belowN
        #print('belowN is winner')
        #print('game value below', game[startN.x, startN.y])
        #print('timeInCell', timeInCell)
        return [cell, bestMove]
    
    elif frontN.x == endN.x and frontN.y == endN.y:
        #timeInCell = game[startN.x, startN.y]
        bestMove=frontN
        #print('FrontN is winner')
        #print('game value front ', game[startN.x, startN.y])
        #print('timeInCell front ', timeInCell)
        return [cell, bestMove]
    
    #Compare values to choose the smallest
    #try blocks to stop crashing if index out of bound
    try:        
        if belowN.x < r  or belowN == previous:
            if game[belowN.x, belowN.y] < timeInCell and belowN != previous:
                timeInCell = game[belowN.x, belowN.y]
                bestMove = belowN
                #print('below time ', timeInCell, 'bestmove coordinates ', bestMove.x, bestMove.y)
    except:
        pass
    
    try:
        if frontN.y < c or frontN == previous:   
            if game[frontN.x, frontN.y] < timeInCell and frontN != previous:
                timeInCell = game[frontN.x, frontN.y]
                bestMove = frontN
                #print('front time', timeInCell, 'bestmove coordinates', bestMove.x, bestMove.y)
    except:
        pass    
    
    #Commented as code keeps looping instead of going forward
    #try:
    #    if aboveN.x != -1 and aboveN.y !=-1 and aboveN != previous:
    #        if game[aboveN.x, aboveN.y] < timeInCell and aboveN != previous:
    #            timeInCell = game[aboveN.x, aboveN.y]
    #             bestMove= aboveN
    #            print('above time ', timeInCell, 'bestMove coordinates', bestMove.x, bestMove.y)
    #except:
    #    pass

    
    return [cell, bestMove]
```
