## Exercise 1 - Logistic Regression

The first part of the exercise is to learn how to manipulate Logistic Regression by using the scikit-learn library.

First of all, we create an artificial dataset with two varibles:
- numbers of hours working on the courserwork: n_hours
- passing the module: passing

Disclaimer: this is of course all synthetic data and bears no relation to the reality.

We will say that students who spend 50 hours working on the CW have a probability to pass which is 0.5.
We will also say that students who spend 90 hours working on the CW have a probablity to pass which is 0.95.
Finally, we will say that students who spend more than 200 hours should take a break.

### Ex 1.1 - Building a Dataset

With these two indications, and using a signoid function with a simple linear combination sigm(ax + b),
we can calculate a and b.

Then, we can sample new points:
- we sample a new n_hours
- we calculate the probability of passing
- we pick a random number between 0 and 1
- if the random number is below te probability of passing, then we assign the label 1, corresponding to passing. Else we assign the label 0.

So we can build datasets X_train, y_train, X_valid, y_valid, X_test, y_test, that we will use to train and evaluate a logistic regression.


```python
import math
import numpy as np
from numpy.random import default_rng
import matplotlib.pyplot as plt

rng = default_rng()

def sigmoid(x):
    
    return 1.0/(1.0+np.exp(-x))

def generate_dataset(N_points, ratio_valid = 0.2, ratio_test = 0.2):
    
    # From solving a system of two equations with two unknowns, we find:
    a = 0.0736
    b = -3.68
    theta = [b, a]
    
    # We generate random values for the number of hours spent.
    # We add X0 = 1 to X so that we can use numpy array operations
    X0 = np.ones(N_points)
    X1 = rng.uniform(0, 100, N_points)
    X = np.vstack([X0, X1]).transpose()
    
    # We can calculate the Probability of passing for each example of number of hours
    P_passing = sigmoid( X.dot(theta) )
    
    # Finally, we can sample a random number to simulate whether a student actually passed of not
    Y = 1 * ( rng.uniform(0, 1, N_points) < P_passing )
    
    # We split the data for training, testing and validation
    X_valid, X_test, X_train = np.split(X, [int(len(X)*ratio_valid), int(len(X)*(ratio_valid+ ratio_test))])
    y_valid, y_test, y_train = np.split(Y, [int(len(X)*ratio_valid), int(len(X)*(ratio_valid+ ratio_test))])
     
    return X_train, y_train, X_test, y_test, X_valid, y_valid

X_train, y_train, X_test, y_test, X_valid, y_valid = generate_dataset(100)
        
plt.scatter(X_train[:,1], y_train, marker = '.')

```

### Ex 1.2 - Implementing Logistic Regression

Similarly to Linear Regression Last week, we can build a Class Logistic Regression that will learn the parameters for us.

Today, you will implement an early-stopping criterion:
- create a validation set
- monitor the history of cost/loss on the validation set
- when the cost on the validation set stops decreasing, then we stop the training

We won't create huge datasets, so we can use Gradient Descent over the full training dataset.


```python
X_train, y_train, X_test, y_test, X_valid, y_valid = generate_dataset(20000)


class LogReg:
    
    def __init__(self, learning_rate = 0.001):
        
        self.params = rng.uniform(-1, 1, 2)
        self.learning_rate = learning_rate
    
    def fit(self, X_train, y_train, X_valid, y_valid, verbose=False, max_epochs=1000):
        
        epoch = 0
        self.loss_history = []
        continue_training = True
        
        y_proba_valid = sigmoid( X_valid.dot(self.params) )
        log_likelihood = np.dot(y_valid.T,  np.log(y_proba_valid)) + np.dot( (1. - y_valid).T, np.log(1-y_proba_valid)) 
        self.loss_history.append(log_likelihood)  
        
        while continue_training:
            
            # Update model parameters with the gradient ascent formulation
            y_proba_train = sigmoid( X_train.dot(self.params) )
            dE_dtheta = (1/len(X_train))*np.dot( X_train.T, y_train-y_proba_train)
            
            self.params += self.learning_rate*dE_dtheta
            
            # Calculate loss on validation set
            y_proba_valid = sigmoid( X_valid.dot(self.params) )
            log_likelihood = y_valid.T.dot( np.log(y_proba_valid)) + (1. - y_valid).T.dot( np.log(1-y_proba_valid) )
        
            # termination condition: when rate of change is below 0.1%
            
            rate = math.fabs(log_likelihood - self.loss_history[-1])/math.fabs(log_likelihood)
#             if rate < 0.00001:
#                 termination_condition = True
#             else:
            termination_condition = False
       
            epoch += 1
            self.loss_history.append(log_likelihood)
            
            if termination_condition or epoch > max_epochs:
                
                continue_training = False
                
            if verbose and epoch%500 == 0:
                print("Epoch: ", epoch, '- loss: ', log_likelihood)
            
    def predict(self, X):
        
        # sigmoid function gives us the probability of the class being one
        proba = sigmoid( X.dot(self.params) )
        
        # We use a decision threshold of 0.5
        y_pred = 1*(proba > 0.5)
        
        return y_pred
    
    def evaluate(self, X_test, y_test):

        y_pred = self.predict(X_test)
        
        return sum(y_pred == y_test)/len(X_test)

    def plot(self, X, y):
        
        # plot the points, their true class (with a different color for each class), and the current sigmoid function.
        X_points_0 = np.ones(1000)
        X_points_1 = np.linspace(0,100,1000)
        X_points = np.vstack([X_points_0, X_points_1]).transpose()
        proba = sigmoid( X_points.dot(self.params) )
        
        plt.plot(X_points[:,1], proba)
        
        y_pred = self.predict(X)
        plt.scatter(X[:,1], y_pred, c=y, marker = '.')
        
        plt.show()
    
    
logreg = LogReg(0.001)

logreg.fit(X_train, y_train, X_valid, y_valid, max_epochs=100000, verbose = True )

print('Accuracy :', logreg.evaluate(X_test, y_test))

plt.plot( logreg.loss_history)



```

```python
logreg.plot(X_test, y_test)
print(logreg.params)
```

Why do we need so many iterations with such small steps?
- When we start our optimization, because the data is not scaled, we start off very far from the center. So the gradient values might be very small.
- The value of number of hours are far greater than the values of the bias (1), so the amplitude of the gradient on one parameter is much higher. We ave to keep the gradient low so that it doesn't diverge.

If we were to use the sklearn version of Logistic Regression, we would benefit from different regularization schemes.

In general, it is often advised to normalize the input data. Different normalization schemes exist that we will see during the Lecture on Neural Networks.

### Ex 2.1 : Blob's burger

We will use sklearn's synthetic dataset generation tools to create a problem where we want to apply logistic regression to a separation of two classes.

Use the Logistic Regression with sklean (with cross validation) to determine the decision boundary and the accuracy of prediction.


```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs( n_samples=1000, n_features=2, centers=((-1, -1), (4, 0)), cluster_std = 2)

plt.scatter( X[:,0], X[:,1], c=y)

```

```python
from sklearn.linear_model import LogisticRegressionCV

ratio_train = 0.8
n_train = int(len(X)*0.8)

X_train, X_test = X[ :n_train], X[ n_train: ]
y_train, y_test = y[ :n_train], y[ n_train: ]

model = LogisticRegressionCV()
model.fit(X_train, y_train)

```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)

plt.scatter( X_test[:,0], X_test[:,1], c=y_pred)
```

## Ex 2.2 : Going circles

In this problem, the two classes belong to two concentric circles. So the decision boundary is a circle.
We can't make a circle with a line. But we can project the input space into a new space of features which will be linear.

Try to find out in your internal memory (brain) or your external memory (internet) the general equation of an ellipsis.
Then propose a transformation of the input space that allows to perform logistic regression.

Measure the accuracy and plot the decision boundary obtained.


```python
from sklearn.datasets import make_circles

X, y = make_circles(1000, noise = 0.2, factor = 0.5)

X = X * (2, 4) + (-2, 5)
plt.scatter( X[:,0], X[:,1], c=y)


```

```python
X_0 = X[:,0]
X_1 = X[:,1]

X_0_squared = X_0**2
X_1_squared = X_1**2

X_0_1 = X_0 * X_1

X_expanded = np.vstack([X_0, X_1, X_0_squared, X_1_squared, X_0_1]).transpose()

ratio_train = 0.8
n_train = int(len(X_expanded)*0.8)


X_train, X_test = X[ :n_train], X[ n_train: ]
X_expanded_train, X_expanded_test = X_expanded[ :n_train], X_expanded[ n_train: ]
y_train, y_test = y[ :n_train], y[ n_train: ]

model = LogisticRegressionCV()
model.fit(X_expanded_train, y_train)

model.score(X_expanded_test, y_test)
```

```python
y_pred = model.predict(X_expanded_test)

plt.scatter( X_test[:,0], X_test[:,1], c=y_pred)
```

## Ex 2.3 : The infamous XOR

Similarly to before, logistic regression alone will not be sufficient to solve the following problem.
As you don't know how many polynomial features you would need, you must try different values of maximum degree of polynomial expansion.

In this last exercise, you can use sklearn PolyomialFeatures function, that will calculate the expansion for you.

```python
from sklearn.datasets import make_blobs

X, y = make_blobs( n_samples=1000, n_features=2, centers=((-2, -2), (-2, 2),(2, 2),(2, -2)), cluster_std = 1)

y[y==2] = 0
y[y==3] = 1

plt.scatter( X[:,0], X[:,1], c=y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.exceptions import ConvergenceWarning

for degree in range(3):
    
    poly = PolynomialFeatures(degree)
    
    X_expanded = poly.fit_transform(X)
    
    X_expanded_train, X_expanded_test = X_expanded[ :n_train], X_expanded[ n_train: ]
    y_train, y_test = y[ :n_train], y[ n_train: ]
    
    model = LogisticRegressionCV(cv = 4)
    model.fit(X_expanded_train, y_train)

    score = model.score(X_expanded_test, y_test)
    
    print('Degree of polynomial: ', degree, '- score : ', score)
```
