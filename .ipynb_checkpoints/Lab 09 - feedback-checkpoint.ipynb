## 1 - XOR classification

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs( n_samples=1000, n_features=2, centers=((-2, -2), (-2, 2),(2, 2),(2, -2)), cluster_std = 1)

y[y==2] = 0
y[y==3] = 1

plt.scatter( X[:,0], X[:,1], c=y)

n_train = 900
X_train, X_test = X[ :n_train,:], X[ n_train:,: ]
y_train, y_test = y[ :n_train], y[ n_train: ]
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

X_train = torch.tensor(X_train).type('torch.FloatTensor')
X_test = torch.tensor(X_test).type('torch.FloatTensor')
y_train = torch.tensor(y_train).type('torch.FloatTensor')
y_test = torch.tensor(y_test).type('torch.FloatTensor')


class XORNet(nn.Module):

    def __init__(self):
        super(XORNet, self).__init__()
        
        self.fc1 = nn.Linear(2, 10) 
        self.fc2 = nn.Linear(10, 1)

    def forward(self, x):
        h1 = F.relu(self.fc1(x))
        o = F.sigmoid(self.fc2(h1))
        return o
    
    def evaluate(self, x, y):
        proba = self.forward(x)
        y_pred = 1.0*(proba > 0.5)
        accuracy = 1.0*torch.sum(y_pred.reshape(-1,) == y)/proba.shape[0]
        return accuracy
        
net = XORNet()
criterion = torch.nn.BCELoss()
optimizer = optim.Adam(net.parameters(), lr=0.01)

# Training for 100 epochs
epochs = 200
net.train()
for e in range(epochs):
    
    optimizer.zero_grad()
    o = net.forward(X_train)
    loss = criterion(o, y_train)
    loss.backward()
    optimizer.step()

    if e%10 == 0:
        print(e, loss)
```

```python
net.eval()
acc = net.evaluate(X_test, y_test)
print(acc)
```
